```python
# importing necessary libraries
import requests
import pandas  as pd
import json

#number of records to fetch
n_records = 500
limit = 100
n_requests = int(n_records/limit)

#API URL
base_url = 'https://globalmart-api.onrender.com'
end_point = '/mentorskool/v1/sales?offset=1&limit=100'
headers = {'access_token': 'fe66583bfe5185048c66571293e0d358'}

# creating a function to fetch data from API
def fetch_data(base_url, end_point, n_requests):
    sales_data = []
    # trying to connect to API
    try:
        print(f'retrieving data...')
        sales_response = requests.get(base_url+end_point, headers= headers)
        sales_response.raise_for_status()
        for i in range(n_requests):
            sales_data.extend(sales_response.json()['data'])
            print(f'{len(sales_data)} records obtained \n')
            end_point = sales_response.json()['next']
    # except handling HTTP error
    except requests.exceptions.HTTPError as e:
        print(f'Error \n{e}')
        return None
    # except handling Request Exception
    except requests.exceptions.RequestException as e:
        print(f'Error \n{e}')
        return None
    return sales_data

# collecting data from API
data = fetch_data(base_url,end_point,n_requests)
data
```

# STEP 2

```python
# normilizing the nested dict data
df= pd.json_normalize(data)
df
```

```python
# cleaning the column names
columns = list(df.columns)
new_columns = [ column.split(".")[-1] for column in columns ]
rename_columns = dict(zip(columns,new_columns))

#renaming the columns
df.rename(columns= rename_columns, inplace=True)

# picking the only necessary columns
necessary_columns = ['id',
                    'sales_amt',
                    'qty',
                    'discount',
                    'profit_amt',
                    'order_id',
                    'order_purchase_date',
                    'order_status',
                    'order_delivered_customer_date',
                    'order_estimated_delivery_date',
                    'product_id',
                    'colors',
                    'category',
                    'customer_id',
                    'customer_name',
                    'VendorID']
df = df[necessary_columns]
df
```
